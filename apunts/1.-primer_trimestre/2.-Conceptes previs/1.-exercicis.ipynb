6. Partint del següent arbre convertiu-lo en un arbre binari de búsqueda.

```python
class Node:
    def __init__(self, valor, fills=[None, None]):
        self.valor = valor
        self.fills = fills
        self.esquerra = fills[0]
        self.dreta = fills[1] if len(self.fills) == 2 else None

    def __str__(self) -> str:
      return f"Node valor: {self.valor}"

    def setEsquerra(self, valor):
      self.fills[0] = self(valor)
      self.esquerra = self.fills[0]

    def setDreta(self, valor):
      self.fills.insert(self(valor))
      self.dreta = self.fills[1]

arbre = Node(1, [
    Node(2, [
        Node(3),
        Node(4, [
            Node(5),
            Node(6)
        ])
    ]),
    Node(7, [
        Node(8),
        Node(9, [Node(10)])
    ]),
])
```

8. Implementa una funció `buscar` que donat un arbre binari de búsqueda i un valor retorni el node que conté aquest valor o `None` si no el troba.

```python

def buscar(arbre, valor):

  if arbre is None:
    return None
  elif arbre.valor is valor:
    return arbre.valor
  major_esquerra = arbre.esquerra is not None and arbre.esquerra.valor >= valor

  if major_esquerra:
    return buscar(arbre.esquerra, valor)
  else:
    return buscar(arbre.dreta, valor)


print(buscar(arbre, 10))
```
