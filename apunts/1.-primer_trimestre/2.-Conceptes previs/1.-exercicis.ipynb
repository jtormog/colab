```python
class Node:
    def __init__(self, valor, fills=None):
        self.valor = valor
        self.fills = fills

    def __str__(self) -> str:
      return f"Node valor: {self.valor}"

    def te_dos_fills(self):
      if self.fills is not None:
        return len(self.fills) == 2
      else:
        return False

    def get_fill_esquerra(self):
      if self.fills is not None:
        return self.fills[0]
      else:
        return False

    def get_fill_dreta(self):
      if self.te_dos_fills():
        return self.fills[1]
      else:
        return False


arbre = Node(1, [
    Node(2, [
        Node(3),
        Node(4, [
            Node(5),
            Node(6)
        ])
    ]),
    Node(7, [
        Node(8),
        Node(9, [Node(10)])
    ]),
])
```

8. Implementa una funció `buscar` que donat un arbre binari de búsqueda i un valor retorni el node que conté aquest valor o `None` si no el troba.

```python

def buscar(arbre, valor):

  te_dos_fills = arbre.te_dos_fills()
  fill_esquerra = arbre.get_fill_esquerra()
  fill_dreta = arbre.get_fill_dreta()

  if valor == arbre.valor:
    return arbre

  if arbre.fills is not None:

    if te_dos_fills:

      if valor < fill_dreta.valor:
        return buscar(fill_esquerra, valor)
      else:
        return buscar(fill_dreta, valor)
    else:
      return buscar(fill_esquerra, valor)


print(buscar(arbre, 5))

```
